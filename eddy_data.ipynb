```python
list_of_param = ["ppt", "t_avg", "t_min", "t_max", "ET_fill"]
for param in list_of_param:
    globals()[f"files_with_column_{param}"] = []


def count_columns(directory, parameter, name_param):
    for filename in os.listdir(directory):
        if filename.endswith('.csv'):
            file_path = os.path.join(directory, filename)
            
            try:
                with open(file_path, 'r', newline='') as csvfile:
                    reader = csv.reader(csvfile)
                    headers = next(reader, None)  # Read the header row
                    
                    if headers and f'{name_param}' in headers:
                        parameter.append(filename)
            except Exception as e:
                print(f"Error reading {filename}: {str(e)}")
    # Print the results
    print(f"Files containing f'{name_param}' column:")
    for file in parameter:
        print(file)

    print(f"\nTotal files with f'{name_param}' column: {len(parameter)}")


def analyze_lists(list1, list2, list3, list4):
    # Convert lists to sets for efficient operations
    set1, set2, set3, set4 = map(set, [list1, list2, list3, list4])
    
    # Find common values across all four lists
    common = set1 & set2 & set3 & set4
    
    # Find values unique to each list
    unique1 = set1 - (set2 | set3 | set4)
    unique2 = set2 - (set1 | set3 | set4)
    unique3 = set3 - (set1 | set2 | set4)
    unique4 = set4 - (set1 | set2 | set3)
    
    # Find values in some but not all lists
    not_common = (set1 | set2 | set3 | set4) - common
    
    return {
        "common": common,
        "unique_to_list1": unique1,
        "unique_to_list2": unique2,
        "unique_to_list3": unique3,
        "unique_to_list4": unique4,
        "in_some_but_not_all": not_common
    }
# /mnt/e/flux_ET_dataset/daily_data_files
directory = 'C:\Users\acer\Desktop\final_code\daily_data_files'
count_columns(directory, files_with_column_ppt, "ppt")
count_columns(directory, files_with_column_t_avg, "t_avg")
count_columns(directory, files_with_column_t_min, "t_min")
count_columns(directory, files_with_column_t_max, "t_max")
count_columns(directory, files_with_column_ET_fill, "ET_fill")

common_value = analyze_lists(
    files_with_column_ppt,
    files_with_column_t_avg,
    files_with_column_t_min,
    files_with_column_t_max
)
 
for key, value in common_value.items():
    print(f"{key}: {value}")

with open(f'C:\Users\acer\Desktop\final_code\data\common_value.pkl', 'wb') as f:
    pickle.dump(common_value, f)
# common_value.pkl
```

```python
def model_4funcs(Q, w, model):
    if model == "Fu":
        ratio = 1 + Q -(1+(Q)**w)**(1/w)
    if model == "Turc-pike":
        ratio = (1 + Q**(-w))**(-1/w)
    if model == "Zhang":
        ratio = (1 + (w * Q)) * (1 + (w * Q) + (Q**(-1)))**(-1)
    if model == "wang-Tang":
        ratio = (1 + Q - math.sqrt(((1 + Q)**2) - (4 * w * (2 - w)) * Q)) / (2 * w * (2 - w))
    return ratio
```

```python
def model_func(x, w):
    return 1 + x - (1 + (x**w))**(1/w)

def aet(ratio, p):
    return ratio * p
```

```python
def calculate_metrics(df,name_col):
    df = df.dropna()
    actual = df['ET_fill']
    predicted = df[name_col]

    rmse = np.sqrt(mean_squared_error(actual, predicted))

    nse = 1 - (np.sum((actual - predicted) ** 2) / np.sum((actual - np.mean(actual)) ** 2))

    nrmse = rmse / (np.max(actual) - np.min(actual))

    r = stats.pearsonr(predicted,actual)[0]
    r2= r**2

    mean_value = sum(predicted) / len(predicted)
    var_value = sum(pow(x-mean_value,2) for x in predicted) / len(predicted)  # variance
    std  = math.sqrt(var_value)  # standard deviation

    return rmse, nse, nrmse, r2, r, std
```
